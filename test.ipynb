```python
import requests
import json
import pandas as pd

url = "https://app.guestmanager.com/api/public/v2/tickets"
headers = {
    'Authorization': 'Token *',
    'Content-Type': 'application/json'
}
params = {
    'page[size]': 50,  # Maximum allowed page size
    'page[number]': 1,  # Start from the first page
    'fields': 'id,first_name,last_name,email,variant,company'
}

all_tickets = []

while True:
    response = requests.get(url, headers=headers, params=params)
    if response.status_code != 200:
        print(f"Failed to fetch data: {response.status_code}")
        break

    try:
        json_data = response.json()
    except ValueError:
        print("Response is not in JSON format")
        print(response.text)  # Print response for debugging
        break

    tickets = json_data.get('data', [])
    if not tickets:
        break  # No more tickets to fetch

    all_tickets.extend(tickets)

    # Check if there's a next page
    next_page = json_data.get('meta', {}).get('page', {}).get('next')
    if not next_page:
        break  # No more pages

    # Move to the next page
    params['page[number]'] += 1

# Convert all tickets data to DataFrame
df = pd.json_normalize(all_tickets)

# Save DataFrame to CSV
df.to_csv('all_tickets.csv', index=False)

print("All tickets data has been saved to all_tickets.csv")

```
